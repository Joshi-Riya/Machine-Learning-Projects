# **T-TEST**

##**HYPOTHESIS 1** : Is the average area_mean for class A and class B is nearly equal or same
## **Null Hypothesis** : There is no significant difference in  the average area_mean
## **Alternate Hypothesis** : There is significant differnce in the average area_mean

```
import random
np.random.seed(0)
Class = dataset['area_mean']
sample_size = 100
class_A = np.random.choice(Class,sample_size,replace=False)
class_B = np.random.choice(Class,sample_size,replace=False)
```

```
print(class_A.mean())
print(class_B.mean())
```

```
from scipy.stats import ttest_ind
stat,p_value = ttest_ind(class_A,class_B)
print("Statistics : ",stat)
print("p_value : ",p_value)
```

```
if p_value > 0.05:
  print("Same distribution i.e we are accepting the null hypothesis")
else:
  print("Different distribution i.e we are rejecting the null hypothesis")
```

# **K-NN CLASSIFICATION MODEL**

# K-NN Model

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 13) 
knn.fit(X_train, y_train)
```

# Predicting The Test Result

```
y_pred = knn.predict(X_test)
np.set_printoptions(precision =2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Confusion Matrix(K-NN)

```
from sklearn.metrics import confusion_matrix
result = confusion_matrix(y_test,y_pred)
print(result)
```

# Model Accuracy

```
from sklearn import metrics
print("ACCURACY: ",metrics.accuracy_score(y_test,y_pred)*100,"%")
```
